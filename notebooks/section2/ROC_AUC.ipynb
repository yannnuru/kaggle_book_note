```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
# 正例、負例のサンプル数
n_sample = 50
# スコアの分布
alpha = 1.0
# 正例の予測スコア
y_pred_pos = 1 - np.random.random_sample(n_sample) * alpha
# 負例の予測スコア
y_pred_neg = np.random.random_sample(n_sample) * alpha
```

```python
fig, axes = plt.subplots(ncols=3, nrows=4, figsize=(3*3, 4*4))

roc_x = []
roc_y = []

for i in range(11):
    thre = 1 - i * 0.1
    fp = (y_pred_neg > thre).sum()
    tp = (y_pred_pos > thre).sum()

    fp_rate = fp / n_sample
    tp_rate = tp / n_sample

    roc_x.append(fp_rate)
    roc_y.append(tp_rate)

    ax = axes[i//3, i%3]
    ax.scatter(np.array(range(0, n_sample)), y_pred_pos, marker='o', label='positive')
    ax.scatter(np.array(range(n_sample, n_sample * 2)), y_pred_neg, marker='x', label='negative')
    ax.hlines(y=thre, xmin=0, xmax=n_sample*2, linestyles='dashed', colors='tab:red', label='threshold')
    ax.set_title(f'fp_rate={fp_rate:.2f}, tp_rate={tp_rate:.2f}')
axes[3,2].scatter(x=roc_x, y=roc_y)
axes[3,2].set_title('roc curve')
axes[3,2].set_xlabel('fp_rate')
axes[3,2].set_ylabel('tp_rate')
axes[0,0].legend()
plt.show()
```
